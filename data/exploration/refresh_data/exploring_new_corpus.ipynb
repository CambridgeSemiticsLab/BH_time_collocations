```python
import collections

from tf.fabric import Fabric
from tf.app import use
```

```python
LOCATION = '/Users/cody/github/BH_time_collocations/data/data/corpus/'
```

```python
tf_fabric = Fabric(LOCATION)
api = tf_fabric.loadAll()
```

```python
app = use('etcbc/bhsa', api=api)
```

```python
F, E, T, L = (getattr(api, l) for l in 'FETL')
```

```python
app.displaySetup(
    condenseType='clause',
    withNodes=True,
    extraFeatures='',
)
```

## All

```python
all_search = app.search("""

clause
/with/
    phrase function=Time
/-/

""")
```

```python
ph_types = collections.Counter()

for res in all_search:
    cl = res[0]
    for ph in L.d(cl, 'phrase'):
        if F.function.v(ph) == 'Time':
            ph_types[F.typ.v(ph)] += 1
            break
```

```python
ph_types.most_common()
```

```python
# other possible time-like phrases

other_search = app.search("""

clause
    phrase function=PreC|Subj
        word lex=JWM/|<RB/|BQR=/|LJLH/|XDC=/CNH/

""")
```
