# Inspect Corpus Edits

```python
import sys
from IPython.display import HTML, display_html
```

```python
from tf.fabric import Fabric
from tf.app import use
```

```python
sys.path.append('/Users/cody/github/BH_time_collocations/data/pipeline/scripts')
from corpus_building.config import THESIS_CORPUS_PARAMS
```

```python
BHSA = '/Users/cody/github/etcbc/bhsa/tf/2021/'
CORPUS = '/Users/cody/github/BH_time_collocations/data/data/corpus/'
```

```python
bhsa_tf = Fabric(BHSA)
bhsa_api = bhsa_tf.loadAll()
bhsa_app = use('etcbc/bhsa', api=bhsa_api)
```

```python
thesis_tf = Fabric(CORPUS)
thesis_api = thesis_tf.loadAll()
thesis_app = use('etcbc/bhsa', api=thesis_api)
```

```python
display_params = {
    'withNodes': True,
    'condenseType': 'verse',
    'hiddenTypes': 'subphrase sentence_atom half_verse'
}

def get_slots_from_edit_action(edit_action):
    """Retrieve slots from edit action."""
    slots = set()
    edge_updates = edit_action.edge_updates or {}
    for node, node_slots in edge_updates.get('oslots', {}).items():
        slots.update(node_slots)
    return slots


def show_edits(edit_actions):
    """Display BHSA and KinghamThesis corpus side-by-side to see changes."""
    for i, edit_action in enumerate(edit_actions, 1):
        slots = sorted(get_slots_from_edit_action(edit_action))
        display_html(HTML('<hr>'))
        display_html(HTML(f'<h3>EDIT {i}</h3>'))
        print(edit_action.description)
        display_html(HTML('<h4>BHSA</h4>'))
        bhsa_app.show([tuple(slots)], **display_params)
        display_html(HTML('<h4>KinghamThesis</h4>'))
        thesis_app.show([tuple(slots)], **display_params)
```
