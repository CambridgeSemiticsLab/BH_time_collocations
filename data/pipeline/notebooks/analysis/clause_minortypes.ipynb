### Various Composite Counts

```python
typfun_ct = pivot_ct(
    df,
    'cl_type2',
    'function',
)

typfun_pr = typfun_ct.div(typfun_ct.sum(1), 0)
typfun_dp = sig.apply_deltaP(typfun_pr, 0, 1)

typgendom_ct = pivot_ct(
    df,
    'cl_type2',
    'gendom',
)

typgendom_pr = typgendom_ct.div(typgendom_ct.sum(1), 0)

typgendom_dp = sig.apply_deltaP(typgendom_ct, 0, 1)
```

```python
top = 7
out.table(
    typfun_ct.iloc[:, :top+1].rename_axis(index='clause type'),
    'type_function_ct',
    caption=f'Clause Type Counts by the Top {top} Time Functions',
    adjustbox=True,
)
```

```python
top = 7
out.table(
    (typfun_pr.iloc[:, :top+1].rename_axis(index='clause type')*100)\
        .round().astype(int).astype(str) + '%'
    ,
    'type_function_pr',
    caption=f'Function Percentages of Clause Type (Top {top} Time Functions)',
    adjustbox=True,
)
```

## Clause Rela

```python
rel_ct = df.cl_rela.value_counts()
rel_pr = rel_ct / rel_ct.sum()

rel_pr
```

```python
rel_both = pd.concat(
    [
        rel_ct,
        rel_pr.mul(100).round().astype(int).astype(str)+'%',
    ],
    1
).set_axis(['count', 'percent'], 1)

out.table(
    rel_both,
    'rel_both',
    caption='Time Adverbial Clause Relation Counts'
)
```

```python
fig, ax = plt.subplots()
rel_pr.plot(kind='bar', ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
out.plot(
    'bar_rel_pr',
)
```

```python
typerel_ct = pivot_ct(
    df,
    'cl_rela',
    'function'
)

typerel_ct = typerel_ct.rename_axis(index='clause relation')

typerel_pr = typerel_ct.div(typerel_ct.sum(1), 0)

typerel_dp = sig.apply_deltaP(typerel_ct, 0, 1)
```

```python
out.table(
    typerel_ct.iloc[:, :top],
    'typerel_ct',
    caption=f'Clause Relationship Counts by Time Functions (Top {top} Functions)',
    adjustbox=True
)
```

```python
typerel_pr.iloc[:, :top]

out.table(
    typerel_pr.iloc[:, :top]\
        .mul(100).round()\
        .astype(int).astype(str)+'%',
    'typerel_pr',
    caption=f'Function Percentages of Clause Relationship (Top {top} Functions)',
    adjustbox=True,
)
```
