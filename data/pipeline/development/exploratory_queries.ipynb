```python
from tf.fabric import Fabric
from tf.app import use
```

```python
CORPUS = '/Users/cody/github/BH_time_collocations/data/data/corpus/'

tf_fabric = Fabric(CORPUS)
tf_api = tf_fabric.loadAll()

app = use('etcbc/bhsa', api=tf_api)

app.displaySetup(**{
    'condenseType': 'clause',
    'withNodes': True,
})

F, E, T, L = (getattr(tf_api, l) for l in 'FETL')
```

```python
time_phrase = """
phrase function=Time
/with/
    =: word lex=B
/or/
    =: word pdp=advb
    <: word lex=B
/-/
"""

time_phrase_set = app.search(time_phrase, shallow=True)
```

```python
len(list(app.search("""

clause
    phrase function=Time
    < phrase function=Time

""", sets={"time_phrase": time_phrase_set})))
```

```python
query = """

tp:time_phrase
/with/
    =: word lex=B
    <: last_word:word st=a prs=absent
    
tp := last_word
/-/

"""


results = sorted(app.search(query, sets={"time_phrase": time_phrase_set}))
```
