# Text Examples

```python
import random
from tf.advanced import options
from tf.app import use
```

```python
bhsa = use('etcbc/bhsa', version='2021')
F, E, T, L = (getattr(bhsa.api, method) for method in "FETL")
```

```python
bhsa.show(
    bhsa.search('''
book book@en=Job
    chapter chapter=20
        verse verse=19
    ''')
)
```

```python
bhsa.displaySetup(
    condenseType='clause',
    extraFeatures='function',
    withNodes=True,
)
```

# Searches

```python
traj_landmark = bhsa.search("""

clause
    phrase function=Loca
        =: word lex=B

""")
```

```python
bhsa.show(
    random.sample(traj_landmark, 10),
)
```
